```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("traffic_data.csv")
df["timestamp"] = pd.to_datetime(df["timestamp"])
```

```python
plt.figure(figsize=(7, 5))
df["vehicle_type"].value_counts().plot(kind="bar")
plt.title("Vehicle Type Distribution")
plt.xlabel("Vehicle Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df.groupby("signal_id")["waiting_time"].mean().plot(kind="bar")
plt.title("Average Waiting Time per Signal")
plt.xlabel("Signal ID")
plt.ylabel("Avg Waiting Time (sec)")
plt.show()
```

```python
plt.figure(figsize=(6, 5))
df["signal_status"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Signal Status Distribution")
plt.ylabel("")  # removes extra label
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df["speed"].plot(kind="hist", bins=10, edgecolor="black")
plt.title("Vehicle Speed Distribution")
plt.xlabel("Speed")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df.groupby(df["timestamp"].dt.second)["vehicle_id"].count().plot(kind="line", marker="o")
plt.title("Vehicles Detected Over Time (by second)")
plt.xlabel("Time (second of minute)")
plt.ylabel("Number of Vehicles")
plt.grid(True)
plt.show()
```
